```python
import glob
import pandas as pd
import csv

merged_files = []
for file in glob.glob("bert-experiment/bert_experiment*.csv"):
    with open(file) as f:
        r = csv.DictReader(f)
        for row in r:
            merged_files.append(row)
            
df = pd.DataFrame(merged_files)
df[[col for col in df.columns if col not in {
"bert_layer",
"bert_mode",
"bert_trainable",
#"epoch"
#"best_epoch"
"training_duration"
    }]] = df[[col for col in df.columns if col not in {
"bert_layer",
"bert_mode",
"bert_trainable",
#"epoch"
#"best_epoch"
"training_duration"
    }]].astype(float)
```
